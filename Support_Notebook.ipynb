```
import numpy as np
import cv2 
from google.colab.patches import cv2_imshow
```

**Initially I will import any image**

```
img = cv2.imread('/content/Lenna_(test_image).png', cv2.IMREAD_REDUCED_COLOR_2)
cv2_imshow(img)
```

#Grayscale Filter

```
#Convert RGB to Greyscale

gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

#Blur Filter

```
# Blur filter 

blur_image = cv2.GaussianBlur(img, (27,27), 0)
cv2_imshow(blur_image)
```

# Sépia Filter

```
# Sepia Filter

img2 = cv2.imread('/content/Lenna_(test_image).png', cv2.IMREAD_REDUCED_COLOR_2)

kernel = np.array([[0.272, 0.534, 0.131],
                   [0.349, 0.686, 0.168],
                   [0.393, 0.769, 0.189]])

sepia_image = cv2.filter2D(img2, -1, kernel)
cv2_imshow(sepia_image)
```

# Sketch filter

```
# Sketch filter

# step 1:
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# step 2:
inv_gray = 255 - gray_image
print('Inverse gray')
cv2_imshow(inv_gray)

# step 3:
blur_image = cv2.GaussianBlur(inv_gray, (25,25), 0)
# Blur applies
print('Blur')
cv2_imshow(blur_image)

# step 4:
sketch_image = cv2.divide(gray_image, 255 - blur_image, scale=256)
print('Final result')
cv2_imshow(sketch_image)
```

#Contour Filter

```
# Canny, Contour: 
blur_image = cv2.GaussianBlur(img, (11,11), 0)
canny_image = cv2.Canny(blur_image, 100, 150)
cv2_imshow(canny_image)
```
